# Прогноз оттока клиентов оператора связи

## Описание проекта

В данном проекте необходимо построить прогноз оттока текущих клиентов для оператора связи, так как маркетологи выяснили, что выгоднее сохранять текущих клиентов, чем привлекать новых. Если выяснится, что пользователь планирует уйти, ему будут предложены промокоды и специальные условия. Прогноз будет строиться на основе собранных персональных данных о некоторых клиентах, информации об их тарифах и договорах.

### Цель проекта

Построение модели прогноза оттока клиентов оператора связи с предельно большим значением метрики AUC-ROC.

### Описание данных

Данные состоят из файлов, полученных из разных источников:

- `contract_new.csv` — информация о договоре;

- `personal_new.csv` — персональные данные клиента;

- `internet_new.csv` — информация об интернет-услугах;

- `phone_new.csv` — информация об услугах телефонии.

Информация о договорах актуальна на 1 февраля 2020.

**Признаки:**

- `customerID` - код клиента;

- `BeginDate` - дата начала пользования услугами;

- `EndDate`- дата окончания пользования услугами;

- `Type` - тип оплаты: месячный, годовой;

- `PaperlessBilling` - электронный платежный документ;

- `PaymentMethod` - способ оплаты;

- `MonthlyCharges` - ежемесячные траты на услуги по договору;

- `TotalCharges` - общие траты;

- `gender` - пол;

- `SeniorCitizen` - наличие пенсионного статуса по возрасту;

- `Partner` - наличие супруга(-и);

- `Dependents` - наличие иждивенцев;

- `InternetService` - подключение к интернету: через телефонную линию (DSL) или оптоволоконный кабель (Fiber optic);

- `OnlineSecurity` - блокировка небезопасных сайтов;

- `OnlineBackup` - облачное хранилище файлов для резервного копирования данных;

- `DeviceProtection` - антивирус;

- `TechSupport` - выделенная линия технической поддержки;

- `StreamingTV` - стриминговое телевидение;

- `StreamingMovies` - каталог фильмов;

- `MultipleLines` - наличие возможности параллельных звонков.

**Целевой признак:**

Будет выделен из столбца `EndDate`.

### Исследовательский анализ данных

```python
#проверим данные на пропуски
df.isna().sum()
```

```python
#заполним пропуски заглушкой "No", тк эти люди не пользуются допуслугами
df = df.fillna('No')
#проверим данные на пропуски
df.isna().sum()
```

Пропуски были устранены заглушкой `No`, так как эти пропуски возникли при присоединении таблиц с дополнительными услугами _internet_ и _phone_, которыми пользуются не все клиенты.

```python
#проверим данные на явные дубликаты
df.duplicated().sum()
```

Явных дубликатов нет.

```python
#поменяем тип данных столбца 'total_charges'
df['total_charges'] = pd.to_numeric(df['total_charges'], errors='coerce')
print(df['total_charges'].isna().sum())
df[df['total_charges'].isna()] 
```

```python
#удалим 11 пропусков в столбце 'total_charges'
df = df.dropna()
#проверим данные на пропуски
df.isna().sum()
```

Заменили тип данных в столбце `total_charges`: был _object_, стал _float64_. Удалили появившиеся при переходе пропуски (_11_).

```python
#напишем функцию, для замены значений столбцов с 'Yes' и 'No' на 1 и 0
'''функция замены значений столбцов 'Yes' и 'No' на 1 и 0
'''
def change_func(value):
    if value == 'No': 
        return 0
    else:
        return 1
```

```python
#выделим целевой признак и добавим его в датафрейм
df['churn_rate'] = df['end_date'].apply(change_func)
df.head()
```

Выделили целевой признак оттока - новый столбец `churn_rate`.

```python
#посмотрим уникальные значения стобца 'end_date'
df['end_date'].value_counts()
```

```python
#заменим значения 'No' на дату выгрузки для подсчета длительности договора
df['end_date'] = df['end_date'].replace('No', '2020-02-01')
#проверим результат
df['end_date'].value_counts()
```

```python
#поменяем тип данных столбца 'end_date'
df['end_date'] = pd.to_datetime(df['end_date'], format='%Y-%m-%d')
```

```python
'''функция подсчета количества дней'''
def num_of_days(row):
    begin = row['begin_date']
    end = row['end_date']
    return (end - begin).days
```

```python
#получим новый столбец 'duration', показывающий длительность договора
df['duration'] = df.apply(num_of_days, axis=1)
df.head()
```

Выделили новый признак - столбец `duration`.

```python
#поменяем значения 1 и 0 на Yes и No в столбцах 'churn_rate' и 'senior_citizen'
df['churn_rate'] = df['churn_rate'].replace({1: 'Yes', 0:'No'}).astype('object')
df['senior_citizen'] = df['senior_citizen'].replace({1: 'Yes', 0:'No'}).astype('object')
```

```python
#создадим переменные для кодирования выборок
features_train_lin = features_train.copy()
features_test_lin = features_test.copy()

#удалим мультиколлинеарные признаки
features_train_lin = features_train_lin.drop(columns=['internet_service', 'streaming_tv', \
                                                      'streaming_movies', 'total_charges']) 
features_test_lin = features_test_lin.drop(columns=['internet_service', 'streaming_tv', \
                                                    'streaming_movies', 'total_charges'])
```

## Обучение разных моделей, выбор лучшей модели

### Обучение LogisticRegression

```python
#обучим модель LogisticRegression
params={'model__C':[.1, 1, 10]}

pipeline = Pipeline([
    ('model', LogisticRegression(solver='liblinear', random_state=22052023, class_weight='balanced'))])

grid_lr = GridSearchCV(pipeline, cv=3, n_jobs=-1, param_grid=params , scoring='roc_auc')

grid_lr.fit(features_train_lin, target_train)


print('Лучшая модель:', grid_lr.best_estimator_)
print('ROC-AUC:', f'{grid_lr.best_score_:.3f}') 
```

В данном пункте была обучена модель _LogisticRegression_, и она получила метрику ROC-AUC: _0.748_. Можно отметить, что это ниже порогового значения.

### Обучение RandomForestClassifier

```python
#обучим модель RandomForestClassifier
params={'model__n_estimators':range(10, 201, 10),
       'model__max_depth':[None] + [i for i in range(2, 7)]}

pipeline = Pipeline([
    ('model', RandomForestClassifier(random_state=22052023, class_weight='balanced'))])

grid_rf = GridSearchCV(pipeline, cv=3, n_jobs=-1, param_grid=params , scoring='roc_auc')

grid_rf.fit(features_train_rf, target_train)


print('Лучшая модель:', grid_rf.best_estimator_)
print('ROC-AUC:', f'{grid_rf.best_score_:.3f}') 
```

В данном пункте была обучена модель _RandomForestClassifier_, и она получила метрику ROC-AUC: _0.821. Можно отметить, что это ниже порогового значения.

### Обучение LGBMClassifier

```python
#обучим модель LGBMClassifier
cat_columns = ['type', 'paperless_billing', 'payment_method', 'gender', 'partner', 'senior_citizen', 'dependents', 
               'internet_service', 'online_security', 'online_backup', 'device_protection', 'tech_support', 'streaming_tv', 
               'streaming_movies', 'multiple_lines']

features_train_b[cat_columns] = features_train_b[cat_columns].astype("category")

params={'model__n_estimators':range(10, 201, 10),
       'model__max_depth':[None] + [i for i in range(2, 7)],
       'model__learning_rate': [0.01, 0.1]}

pipeline = Pipeline([
    ('model', lgb.LGBMClassifier(random_state=22052023, class_weight='balanced'))])

grid_ltb = GridSearchCV(pipeline, cv=3, n_jobs=-1, param_grid=params , scoring='roc_auc')

grid_ltb.fit(features_train_b, target_train)


print('Лучшая модель:', grid_ltb.best_estimator_)
print('ROC-AUC:', f'{grid_ltb.best_score_:.3f}')
```

В данном пункте была обучена модель _LGBMClassifier_, и она получила метрику ROC-AUC: _0.856_. Можно отметить, что это выше порогового значения.

### Обучение CatBoostClassifier

```python
#обучим модель CatBoostClassifier
cat_features = ['type', 'paperless_billing', 'payment_method', 'gender', 'partner', 'senior_citizen', 'dependents', 
                'internet_service', 'online_security', 'online_backup', 'device_protection', 'tech_support', 'streaming_tv', 
               'streaming_movies', 'multiple_lines']

params={'model__depth':[4, 6, 8],
        'model__learning_rate': [0.01, 0.1]}

pipeline = Pipeline([
    ('model', CatBoostClassifier(cat_features=cat_features, random_state=22052023, auto_class_weights='Balanced'))])

grid_cb = GridSearchCV(pipeline, cv=3, n_jobs=-1, param_grid=params , scoring='roc_auc', verbose=False)

grid_cb.fit(features_train_cb, target_train, model__verbose=False)


print('Лучшая модель:', grid_cb.best_estimator_, '\n', grid_cb.best_params_)
print('ROC-AUC:', f'{grid_cb.best_score_:.3f}')
```

В данном пункте была обучена модель _CatBoostClassifier_, и она получила метрику ROC-AUC: _0.887_. Можно отметить, что это выше порогового значения.

### Выбор лучшей модели

```python
#сделаем таблицу по полученным метрикам для выбора лучшей модели
model_sel = {
    'ROC-AUC': [f'{grid_lr.best_score_:.3f}', f'{grid_rf.best_score_:.3f}', \
                f'{grid_ltb.best_score_:.3f}', f'{grid_cb.best_score_:.3f}']
}

model_selection = pd.DataFrame(model_sel, index=['LogisticRegression', 'RandomForestClassifier',\
                                                 'LGBMClassifier', 'CatBoostClassifier'], columns=['ROC-AUC'])
model_selection
```

Самой большой метрикой _ROC-AUC_, превышающей пороговое значение, обладает модель _CatBoostClassifier_ с параметрами  _'depth': 4, 'learning_rate': 0.1, random_state=22052023, auto_class_weights='Balanced'_. Эту лучшую модель проверим на тестовой выборке.
